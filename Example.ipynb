```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

from ridgeplot import ridgeplot
from colors import ColorEncoder, ColorPalette
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {}
for i in range(10):
    data['data_{}'.format(i)] = np.random.randn(100) * (i+1)

fig = plt.figure()
ax = fig.add_subplot(111)
ridgeplot(ax, 
          data, 
          xlim=(-20,20), 
          label_size=15)
```

## Different colors ##

```python
ColorPalette.simpsons
```

```python
color_encoder = ColorEncoder()
color_encoder.fit(
    categories = list(data.keys()),
    colors = ColorPalette.simpsons.value,
)
color_encoder.encoder
```

```python
line_colors = color_encoder.transform(list(data.keys()))
line_colors
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ridgeplot(ax, 
          data, 
          xlim=(-20,20), 
          label_size=10, 
          line_colors=line_colors, 
          fill_colors=line_colors)
```

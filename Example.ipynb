```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

from ridgeplot import ridgeplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {}
for i in range(10):
    data['data_{}'.format(i)] = np.random.randn(100) * (i+1)

fig = plt.figure()
ax = fig.add_subplot(111)
ridgeplot(ax, 
          data, 
          xlim=(-20,20), 
          label_size=15)
```

## Different colors ##

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ridgeplot(ax, 
          data, 
          xlim=(-20,20), 
          label_size=10, 
          line_colors=['#1ddd%s' %str(i*10).rjust(2,'0') for i in range(len(data))], 
          fill_colors=['#6350%s' %str(i*10).rjust(2,'0') for i in range(len(data))])
```
